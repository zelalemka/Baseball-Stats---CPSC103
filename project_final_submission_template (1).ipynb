<font color="blue">
We are designing a data definition that represents a baseball pitcher's performance.

Their performance is determined by the number of barrels (int), their meatball percentage (float), their number of pitches thrown (int), and their xwOBA (float). 
    
Note again that a 'meatball' is defined as a pitch that is (in theory) incredibly easy for a professional batter to hit a home run on. Pitchers that throw a higher percentage of meatballs should be less successful.

Our desired graph is going to be comparing the number of barrels per meatball to xwOBA rating - in order to get the number of barrels per meatball, we need to multiply the number of pitches thrown and the meatball percentage to calculate the number of meatballs thrown, then divide that by the player's total number of barrels. That final number is the barrels per meatball, otherwise known as meatball quotient.

Note that both meatball quotient and number of meatballs thrown are already present in the original csv file, but will be calculated again for this project as computation. 

In order for this data definition to work well with all different sizes of data, we are using the arbitrary sized data definition, and since the values are all related to eachother, we will be using the compound data template
</font>

```python
from cs103 import *
from typing import NamedTuple, List
import csv

##################
# Data Definitions
PitcherStats = NamedTuple('PitcherStats', [('barrels', int),
                                           ('meatball_percent', float),
                                           ('pitches', int),
                                           ('xwOBA', float)])                                        
# interp. number of times the batter hits it 'perfectly, percentage of meatball throws,
# number of times said player pitched, and that player's xWOBA (performance metric)
PS1 = PitcherStats(50, 4.2, 2726, 0.372)
PS2 = PitcherStats(58, 5.4, 2674, 0.356)
PS3 = PitcherStats(5, 10.4, 970, 0.287)
PS4 = PitcherStats(5, 9.4, 1129, 0.263)
PS5 = PitcherStats(3, 6.6, 1042, 0.23)
                          
@typecheck
def fn_for_pitcher_stats(ps: PitcherStats) -> ...: # template based on Compound
    return ...(ps.barrels, # int in range [0, ...)
               ps.meatball_percent, # float in range [0, 100]
               ps.pitches, #int in range [0, ...)
               ps.xwOBA) # float in range [0, 1] ??????

# List[PitcherStats]
# interp. a list of PitcherStats

LOPS0 = []
L0PS1 = [PS1, PS2]
L0PS2 = [PitcherStats(10, 2.3, 1000, 0.278), PitcherStats(20, 3.3, 2000, 0.333)]

@typecheck
def fn_for_lop(lops: List[PitcherStats]) -> ...: # template based on arbitrary-sized and the 
    # description of the accumulator          # reference rule
    acc = ...      # type: ...
    for ps in lops:
        acc = ...(fn_for_pitcher_stats(ps), acc)
    return ...(acc)
```

```python
###########
# Pre-Plotting Functions

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename and produces a scatterplot comparing the meatball quotient to the xwOBA 
    values of the pitchers in the file.
    """
    # Template from HtDAP, based on function composition 
    return plot_data(read(filename)) 
    
    

@typecheck
def read(filename: str) -> List[PitcherStats]:
    """    
    reads information from the specified file  and returns a list of pitcher stats that determine performance
    """
    #return []  #stub
    # Template from HtDAP
    # lops contains the result so far
    lops = [] # type: List[PitcherStats]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip first (and only) header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            ps = PitcherStats(parse_int(row[2]), parse_float(row[3]), parse_int(row[4]) ,
                              parse_float(row[7]))
            lops.append(ps)
    
    return lops

@typecheck
def meatball_quotient(ps: PitcherStats) -> float:
    """
    returns meatball quotient by multiplying pitches by meatball percent (divided by 100 to create 
    the percentage) and then dividing that value by the number of barrels to get meatball per barrel AKA meatball 
    quotient. 
    """
    # return 0.0 #stub
    # template PitcherStats
    return ((ps.meatball_percent / 100) * ps.pitches) / ps.barrels

@typecheck
def mq_list(lops: List[PitcherStats]) -> List[float]: 
    """
    returns a list containing the calculated meatball quotient for List of PitcherStats
    """
    #return [] #stub
    #template from List[PitcherStats]
    # acc contains the result do far
    list = [] # type: List[PitcherStats]
    for ps in lops:
        list.append(((ps.meatball_percent / 100) * ps.pitches) / ps.barrels)
    return list

@typecheck 
def xwoba_list(lops: List[PitcherStats]) -> List[float]:
    """
    returns a list containing the xwoba values needed to be plotted on the y axis
    """
    xwoba = []
    for ps in lops: 
        xwoba.append(ps.xwOBA)
    return xwoba
    
start_testing()

# Examples and tests for main
expect(main('meatball_quotient3.csv'), None)

summary()

start_testing()

# Examples and tests for read
expect(read('meatball_quotient_test1.csv'), [PS1, PS2])
expect(read('meatball_quotient_test2.csv'), [PS3, PS4, PS5])
summary()


start_testing()
#Examples and tests for Meatball Quotient
expect(meatball_quotient(PS3), 20.176000000000002)
expect(meatball_quotient(PS4), 21.2252)
expect(meatball_quotient(PitcherStats(10, 2.3, 1000, 0.278)), 2.3)
expect(meatball_quotient(PitcherStats(20, 3.3, 2000, 0.333)), 3.3)
       

summary()

start_testing()

# Examples and tests for List of meatball quotient 
expect(mq_list([]), [])
expect(mq_list([PS1, PS2]), [2.28984, 2.489586206896552])
expect(mq_list([PS3, PS4]), [20.176000000000002, 21.2252])
expect(mq_list([PitcherStats(10, 2.3, 1000, 0.278),
               PitcherStats(20, 3.3, 2000, 0.333)]), [2.3, 3.3])


summary()

start_testing()
#Examples and tests for list of xwoba values
expect(xwoba_list([]), [])
expect(xwoba_list([PS1, PS2]), [0.372, 0.356])
expect(xwoba_list([PS3, PS4]), [0.287, 0.263])
expect(xwoba_list([PitcherStats(10, 2.3, 1000, 0.278),
               PitcherStats(20, 3.3, 2000, 0.333)]), [0.278, 0.333])

summary()

```

```python
##### Plotting functions
import matplotlib.pyplot as plt

@typecheck
def plot_data(lops: List[PitcherStats]) -> None:
    """
    display a scatterplot of meatball quotient vs. xwOBA.
    
    """
    #return None #stub
    # Template based on visualization
    

    # set the labels for the axes
    plt.xlabel('Meatball quotient (Barrels per meatball)')
    plt.ylabel('xwOBA')
    plt.title('Relationship between meatball quotient and pitcher performance by xwOBA')

    # range for the axes
    # [x-min, x-max, y-min, y-max]
    plt.axis([0,23,0.2,0.45])

    # create the scatterplot, with markers that are red (c='r') and triangular (marker='^')
    x_values = mq_list(lops)
    y_values = xwoba_list(lops)
    plt.scatter(x_values, y_values, marker='^', c='r', alpha = 0.5)

    # show the plot
    plt.show()
    
    return None



start_testing()

# Should display an empty graph with x-axis "Meatball quotient", y-axis "xwOBA", and
# title "Relationship between meatball quotient and pitcher performance by xwOBA".
expect(plot_data([]), # xwoba
       None) 

summary()

# Should display a graph labelled much the same above, but now with content
expect(plot_data([PitcherStats(10, 2.3, 1000, 0.278),
               PitcherStats(20, 3.3, 2000, 0.333)]),  
        None)     

summary()

#Displays graph with 5 Pitchers with extreme meatball quotients
expect(plot_data([PS1, PS2, PS3, PS4, PS5]),  
        None)     

summary()
```
